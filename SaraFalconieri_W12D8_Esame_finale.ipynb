Esercizio 2: Si chiede di trovare, per ogni continente:
a. il numero di casi fin dall'inizio della pandemia
b. la percentuale rispetto al totale mondiale del numero di casi

```python
# Filtriamo solo i dati dei continenti (escludendo i singoli paesi)
dati_continenti = df[df['continent'].notna()]

# Sommiamo i casi totali per ogni continente (valore massimo registrato)
totale_casi_continente = dati_continenti.groupby('continent')['total_cases'].max().reset_index()

# Calcoliamo il totale dei casi a livello mondiale
totale_casi_mondiali = totale_casi_continente['total_cases'].sum()

# Calcoliamo la percentuale di casi per ogni continente rispetto al totale mondiale
totale_casi_continente['percentuale_sul_totale'] = (totale_casi_continente['total_cases'] / totale_casi_mondiali) * 100

# Stampiamo il risultato
print(totale_casi_continente)
```

# Esercizio 3: Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati: a. l'evoluzione del casi totali dall'inizio alla fine dell'anno b. il numero di nuovi casi rispetto alla data; poiché i nuovi casi vengono registrati settimanalmente, filtrare via i dati che non hanno misurazioni

```python
import matplotlib.pyplot as plt
import math

# Filtriamo i dati per l'Italia nel 2022
dati_italia_2022 = df[(df['location'] == 'Italy') & (df['date'].between('2022-01-01', '2022-12-31'))]

# Convertiamo la colonna 'date' in formato datetime
dati_italia_2022['date'] = pd.to_datetime(dati_italia_2022['date'])

# Ordiniamo i dati per data
dati_italia_2022 = dati_italia_2022.sort_values(by='date')

# Arrotondiamo il totale dei casi con math.ceil
dati_italia_2022['total_cases'] = dati_italia_2022['total_cases'].apply(lambda x: math.ceil(x) if not math.isnan(x) else 0)

# Grafico: Evoluzione dei casi totali
plt.figure(figsize=(10, 5))
plt.plot(dati_italia_2022['date'], dati_italia_2022['total_cases'], marker='o', linestyle='-', label="Casi Totali")
plt.xlabel('Data')
plt.ylabel('Casi Totali')
plt.title('Evoluzione dei Casi Totali in Italia nel 2022')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()

# Filtriamo i dati con nuovi casi settimanali registrati (senza valori NaN)
dati_settimanali = dati_italia_2022.dropna(subset=['new_cases'])

# Arrotondiamo i nuovi casi con math.ceil
dati_settimanali['new_cases'] = dati_settimanali['new_cases'].apply(lambda x: math.ceil(x) if not math.isnan(x) else 0)

# Grafico: Nuovi casi rispetto alla data
plt.figure(figsize=(10, 5))
plt.bar(dati_settimanali['date'], dati_settimanali['new_cases'], color='orange', label="Nuovi Casi")
plt.xlabel('Data')
plt.ylabel('Nuovi Casi')
plt.title('Nuovi Casi in Italia nel 2022 (Solo Dati Settimanali)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend()
plt.show()
```

# Esercizio 4: Riguardo le nazioni di Italia, Germania e Francia mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) da maggio 2022 (incluso) ad aprile 2023 (incluso)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtriamo i dati per Italia, Germania e Francia tra maggio 2022 e aprile 2023
dati_europei = df[
    (df['location'].isin(['Italy', 'Germany', 'France'])) &
    (df['date'].between('2022-05-01', '2023-04-30'))
]

# Convertiamo la colonna 'date' in formato datetime
dati_europei['date'] = pd.to_datetime(dati_europei['date'])

# Selezioniamo solo le colonne necessarie e rimuoviamo i valori nulli
dati_icu = dati_europei[['location', 'date', 'icu_patients']].dropna()

# Creiamo un boxplot per confrontare i pazienti in terapia intensiva tra i tre paesi
plt.figure(figsize=(8, 6))
sns.boxplot(x='location', y='icu_patients', data=dati_icu)
plt.xlabel('Paese')
plt.ylabel('Numero di Pazienti in Terapia Intensiva (ICU)')
plt.title('Confronto dei pazienti in terapia intensiva tra Italia, Germania e Francia (Mag 2022 - Apr 2023)')
plt.grid()

# Stampiamo il grafico
plt.show()
```

Commento:
Il boxplot rappresenta la distribuzione del numero di pazienti in terapia intensiva nei paesi Francia, Germania, Italia. I Box descrivono la variazione interquartile, dove cade il 50% dei dati. Le linee mostrano la dispersione dei valori più estremi. I punti fuori dal box sono outlier.

Principali rilevanze:

*   Germania e Francia hanno avuto più pazienti in terapia intensiva rispetto all'Italia: la mediana della Germania e della Francia è più alta rispetto a quella dell'Italia; la scatola della Germania e della Francia è posizionata su valori più alti rispetto all'Italia.
*   La Germania mostra una maggiore variabilità rispetto alla Francia: la scatola della Germania è più ampia, il che indica una maggiore fluttuazione nel numero di pazienti in terapia intensiva; anche i "baffi" della Germania sono più lunghi, indicando un range più esteso di dati.
*   L'Italia ha avuto valori di terapia intensiva costantemente più bassi: Il range interquartile dell'Italia (la scatola) è più compatto e su valori molto più bassi rispetto agli altri due paesi; questo suggerisce che il numero di pazienti in terapia intensiva è rimasto più stabile e inferiore rispetto a Germania e Francia.

Quindi:
- La Germania ha avuto il numero più alto di pazienti in terapia intensiva e con maggiore variabilità;
- La Francia ha avuto un numero di pazienti simile alla Germania, ma con meno variabilità;
- L'Italia ha avuto meno pazienti in terapia intensiva ed è rimasta più stabile nel tempo.

# Esercizio 5: Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2023: a. mostrare la somma dei pazienti ospitalizzati per ognuna b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o meno

```python
import pandas as pd

# Caricare il dataset
df = pd.read_csv('/content/owid-covid-data.csv')

# Convertire la colonna 'date' in formato datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Filtrare i dati per le nazioni di interesse (Italia, Germania, Francia, Spagna) nel 2023
dati_2023 = df[(df['location'].isin(['Italy', 'Germany', 'France', 'Spain'])) &
               (df['date'] >= '2023-01-01') & (df['date'] <= '2023-12-31')]

# Selezionare solo le colonne necessarie
dati_ospedalizzati = dati_2023[['location', 'hosp_patients']]

# Contare il numero di valori nulli nella colonna 'hosp_patients'
valori_nulli = dati_ospedalizzati['hosp_patients'].isna().sum()

# Sommare il numero di pazienti ospedalizzati per ogni nazione, ignorando i valori nulli
somma_ospedalizzati = dati_ospedalizzati.groupby('location', as_index=False)['hosp_patients'].sum()

# Mostrare i risultati
print("🔹 Somma dei pazienti ospedalizzati nel 2023 per nazione:")
print(somma_ospedalizzati)

# Stampiamo il numero di valori nulli
print(f"\n🔹 Numero di valori nulli nella colonna 'hosp_patients': {valori_nulli}")

```

Osservazioni:
- Francia e Italia hanno registrato il maggior numero di pazienti ospedalizzati nel 2023.
- La Germania ha un valore pari a zero, il che può significare due cose:
  - I dati sulla Germania potrebbero essere completamente mancanti nel dataset.
  - Potrebbero esserci problemi nella raccolta dei dati per questo paese.
- Ci sono 843 valori nulli nella colonna 'hosp_patients', il che significa che alcune date mancano di registrazioni.

Se i valori mancanti sono pochi e sparsi nel tempo, possiamo sostituirli con la media mobile o un'interpolazione per ottenere un trend più continuo.
Se i dati sono mancanti per lunghi periodi o intere nazioni, è meglio non sostituirli, perché significherebbe inventare dati che potrebbero non riflettere la realtà.
In tal caso, la Germania ha valori tutti nulli, non possiamo sostituirli in modo affidabile per cui dobbiamo escluderla dall'analisi.


A questo punto eslcudiamo la Germania e rappresentiamo graficamente il trend per Italia, Spagna e Francia.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Carico il dataset
df = pd.read_csv('/content/owid-covid-data.csv')

# Converto la colonna 'date' in formato datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Filtro i dati per Francia, Italia e Spagna nel 2023 (escludendo la Germania)
dati_2023 = df[(df['location'].isin(['France', 'Italy', 'Spain'])) &
               (df['date'] >= '2023-01-01') & (df['date'] <= '2023-12-31')]

# Rimuovo righe con valori nulli nella colonna 'hosp_patients'
dati_validi = dati_2023.dropna(subset=['hosp_patients'])

# Creiamo il grafico del trend temporale dei pazienti ospedalizzati
plt.figure(figsize=(12, 6))

# Creiamo una linea per ogni nazione
for paese in ['France', 'Italy', 'Spain']:
    subset = dati_validi[dati_validi['location'] == paese]
    plt.plot(subset['date'], subset['hosp_patients'], label=paese)

# Personalizziamo il grafico
plt.xlabel('Data')
plt.ylabel('Numero di Pazienti Ospedalizzati')
plt.title('Trend dei Pazienti Ospedalizzati in Francia, Italia e Spagna (2023)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Stampiamo il grafico
plt.show()
```
